# **Gráficos**

```r
# Gráfico de dispersão (scatterplot)
x <- c(1, 2, 3, 4, 5)
y <- c(2, 3, 5, 6, 7)
plot(x, y, main="Gráfico de Dispersão", xlab="X", ylab="Y")
```

# **Manipulação de Dados**

```r
# Criando um dataframe
nome <- c("João", "Maria", "Pedro", "Ana", "I1", "I2","I3","I4","I5")
idade <- c(25, 30, 35, 40,25, 30, 35, 40,23)
salario <- c(3000, 3500, 4000, 4500,3000, 3500, 4000, 4500, 4500)
```

```r
dados <- data.frame(nome, idade, salario)
```

```r
dados
```

```r
# Exibir os primeiros registros
head(dados)
```

```r
# Exibir os últimos registros
tail(dados)
```

```r
# Filtrar dados
dados_filtrados <- subset(dados, salario > 3500)
print(dados_filtrados)
```

```r
dados_filtrados$nome
```

```r
# Ordenar dados
dados_ordenados <- dados[order(dados$salario), ]
print(dados_ordenados)
```

# **Exemplos de Data Mining**

```r
# Exemplo 1: Clusterização K-means
library(cluster)
dados <- data.frame(x = rnorm(100), y = rnorm(100))
kmeans_result <- kmeans(dados, centers=3)
print(kmeans_result)
```

```r
# Exemplo 2: Regressão Linear
dados <- data.frame(x = 1:100, y = 2*(1:100) + rnorm(100))
regressao <- lm(y ~ x, data=dados)
summary(regressao)
```

```r
# Exemplo 3: Árvore de Decisão
library(rpart)
dados <- data.frame(x = runif(100), y = as.factor(ifelse(runif(100) > 0.5, "A", "B")))
arvore <- rpart(y ~ x, data=dados, method="class")
print(arvore)
```
