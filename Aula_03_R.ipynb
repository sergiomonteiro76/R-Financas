# **Dataframe**

**Distribuição Normal**

```r
x = rnorm(1000, mean = 10, sd = 2)
x
```

```r
x[1]
```

```r
media <- mean(x)
print(paste("Média:", media))
# Desvio Padrão
desvio_padrao <- sd(x)
print(paste("Desvio Padrão:", desvio_padrao))
```

```r
# Gerando dados seguindo uma distribuição normal
dados_normal <- data.frame(x)

# Histograma dos dados
hist(dados_normal$x, main = "Distribuição Normal", xlab = "Valor", ylab = "Frequência")
```

**Distribuição de Poisson**

```r
# Gerando dados seguindo uma distribuição de Poisson
dados = rpois(1000, lambda = 5)
```

```r
dados_poisson <- data.frame(dados)

# Histograma dos dados
hist(dados_poisson$dados, main = "Distribuição de Poisson", xlab = "Valor", ylab = "Frequência")
```

**Exponencial Negativa**

```r
# Gerando dados seguindo uma distribuição exponencial negativa
w = rexp(1000, rate = 0.2)
dados_exp_neg <- data.frame(w)
```

```r
# Histograma dos dados
hist(dados_exp_neg$w, main = "Exponencial Negativa", xlab = "Valor", ylab = "Frequência")
```

**Filtragem de Dados**

```r
# Criando um dataframe com dados aleatórios
dados <- data.frame(x = rnorm(100), y = rnorm(100))

# Filtrando dados onde x é maior que 0
dados_filtrados <- subset(dados, x > 0)

# Scatterplot dos dados filtrados
#plot(dados_filtrados$x, dados_filtrados$y, main = "Dados Filtrados", xlab = "X", ylab = "Y")
plot(dados$x, dados$y, main = "Dados Filtrados", xlab = "X", ylab = "Y")
```

**Ordenação de Dados**

```r
# Ordenando dados em ordem crescente
dados_ordenados <- dados[order(dados$x), ]

# Plotando dados ordenados
plot(dados_ordenados$x, dados_ordenados$y, main = "Dados Ordenados", xlab = "X", ylab = "Y")
```

**Adição de Colunas**

```r
# Adicionando uma nova coluna com valores aleatórios ao dataframe existente
dados$z <- rnorm(100)
dados
```

**Agregação de Dados**

```r
letters
```

```r
letters[1:3]
```

```r
letters[20:26]
```

```r
grupo = rep(letters[1:3], each = 10)
print('grupo')
grupo
```

```r
valor = rnorm(30)
print('valor')
valor
```

```r
# Criando um dataframe com grupos e valores
dados_agregados <- data.frame(grupo, valor)

# Calculando a média por grupo
media_por_grupo <- aggregate(valor ~ grupo, data = dados_agregados, FUN = mean)
media_por_grupo
```

**Junção de Dataframes**

```r
# Criando dois dataframes
df1 <- data.frame(id = 1:5, valor1 = rnorm(5))
df2 <- data.frame(id = 3:7, valor2 = rnorm(5))

# Inner join
df_joined <- merge(df1, df2, by = "id", all = TRUE)
df_joined
```

**Remoção de Duplicatas**

```r
dados_duplicados <- data.frame(x = c(1, 2, 3, 3, 1, 3), y = c("a", "b", "c", "c", "a", "c"))
print('dados_duplicados')
dados_duplicados
# Removendo duplicatas
dados_sem_duplicatas <- unique(dados_duplicados)
print('dados_sem_duplicatas')
dados_sem_duplicatas
```

# **Geração e Manipulação de Dados**

```r
#gerar sequência
x <- 1:5
print(x)
```

```r
# [1] 1 2 3 4 5
#chamar função que existe
x <- cumsum(x)
print(x)
# [1] 1 3 6 10 15
```

```r
#formas equivalentes de gerar sequência
x <- seq_len(25)
print(x)
y <- 1:25
print(y)
```

```r
#somar os valores de x acumulativamente
x <- cumsum(x)
print(x)
#imprimir x no gráfico de barras
barplot(x)
```
